```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

plt.style.use('seaborn')
```

```python
# Getting the data ready
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease.head(20)
```

```python
# creating the label and targets for model evaluation 
X = heart_disease.drop('target', axis=1)
y = heart_disease['target']
```

```python
# using the randomforest classififier algorithm
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

clf.get_params()
```

```python
# splitting and fitting the model 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train
```

```python
# fiting the data to a model 
clf.fit(X_train, y_train)
```

```python
X_test
```

```python
y_preds = clf.predict(X_test)
y_preds
```

#### Evaluate the model on the training and test data

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
sklearn.__version__
```

```python
accuracy_score(y_test, y_preds)
```

####  Improve a Model

```python
np.random.seed(42)

for i in range(10, 110, 10):
    print(f'Tring model on {i} estimators...')
    model = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f'Test set accuracy {model.score(X_test, y_test):.2f}%')
print(f'Optimun estimator parameter : {model.score(X_test, y_test).max()}')
```

```python
# 6 save a model and load it 
import pickle
data = model
pickle.dump(data, open('data.pkl', 'wb'))
```

```python
# loading pickle 
model = pickle.load(open('data.pkl', 'rb'))
model.score(X_test, y_test)
```
