##  Anatomy Of A Matplotlib Figure

#####  A typical matplotlib plot is made up of a figure and numerous axes  which combine together to produce a graph, bar chart, pie chart etc.
It can be customized by adding a title, legend, x & y labels and so on 

###  Scatter and Bar Plot
##### These are some common medium of visualization 

```python
x = np.linspace(0, 10, 100)
x
```

```python
# Line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
ax.set(title='Line Plot',
      xlabel='Number of X', 
      ylabel='Square of X'
      )
plt.show()
```

```python
# scatter Plot
fig, ax =plt.subplots()
ax.scatter(x, np.exp(x))
ax.set(title='Scatter Plot',
      xlabel='Number of X', 
      ylabel='Square of X'
      )
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# plot from a dictionary

nut_butter_price = {'Almond Butter': 10,
                   'Peanut Butter': 8,
                   'Cashew Butter': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values(),)
plt.show()
```

```python
nut_butter_price = {'Almond Butter': 10,
                   'Peanut Butter': 8,
                   'Cashew Butter': 12}
fig, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()), nut_butter_price.values(),)
plt.show()
```

####  New dataset : heart disease

```python
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease
```

```python
#create a histogram
heart_disease['age'].plot.hist()
plt.show()
```

```python
# working with patients with age > 50 
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# scatter plot of the age and chol level using the API method
over_50.plot('age', 'chol', kind='scatter',
             c='target', figsize=(10, 5),
            )
plt.show()
```

```python
# using the OO method
fig,ax = plt.subplots(figsize=(10, 5))
scatter = over_50.plot(x='age', y='chol',
             kind='scatter', c='target',
             ax=ax, cmap='jet' 
            )
ax.axhline(over_50['chol'].mean(), linestyle='--')
plt.show()
```

```python
over_50.head()
```

```python
# plot of the age, chol and thalach
fig, ax = plt.subplots(2,1, figsize=(10, 10), sharex=True)
ax[0].set_xlim([50, 75])
#plot for cholesterol against age, ax[0]
scatter = ax[0].scatter('age', 'chol', data=over_50,
                        c='target', cmap='jet')
ax[0].axhline(over_50['chol'].mean(), linestyle='--', c='green')
ax[0].legend(*scatter.legend_elements(), title='target')
ax[0].set(title='Heart Disease and Cholesterol level',
         ylabel='Cholesterol')

# plot of thalach against age
scatter2 = ax[1].scatter(x='age', y='thalach', data=over_50,
                       c='target', cmap='jet')
ax[1].set(title='Heart Disease and Max heart',
          xlabel='Age', ylabel='Thalach')
ax[1].legend(*scatter2.legend_elements(), title='target')
ax[1].axhline(over_50['thalach'].mean(), linestyle=':', color='blue')

plt.suptitle('Heart Disease Analysis', fontsize='20', fontweight='bold')
plt.show()
```
