####  Creating New data

```python
seat_column = pd.Series([5,5,5,5,5])

car_sale['Seats'] = seat_column
car_sale
```

```python
car_sale['Seats'].fillna(5, inplace=True)
car_sale
```

```python
fuel_economy = [7.5, 9.2, 5.0, 9.8, 3.2, 9.1, 8.2, 7.3, 1.2, 2.3]
car_sale['fuel_per_100km'] = fuel_economy
car_sale
```

```python
car_sale['Total fuel used (L)'] = (car_sale['Odometer (KM)']/100) * car_sale.fuel_per_100km
car_sale['Total fuel used'] = (car_sale['Odometer (KM)']/100) * car_sale.fuel_per_100km
```

```python
car_sale
```

```python
#creating a column from a single value
car_sale['Number of wheels'] = 4
car_sale
```

```python
car_sale['Passed road safty'] = True
car_sale
```

```python
# Removing a column using drop
car_sale.drop('Total fuel used', axis=1, inplace=True)
car_sale
```

```python
car_sale_shuffled = car_sale.sample(frac=1)
car_sale_shuffled
```

```python
car_sale_shuffled.reset_index(drop=True, inplace=True)
car_sale_shuffled
```
