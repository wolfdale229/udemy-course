#### Using a pandas data's frame


```python
car_sales = pd.read_csv('data/car-sales.csv')
car_sales.head(10)
```

```python
# refine the Price values, change from str onto integer

car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '').astype(int)
#divide Price by 100
car_sales['Price'] = car_sales['Price']/100
```

```python
car_sales
```

```python
plt.plot(car_sales.Price)
plt.show()
```

```python
# adding a total sales column
car_sales['Total sales ($)'] = car_sales['Price'].cumsum()
car_sales
```

```python
# adding a date column
car_sales['Sale Date'] = pd.date_range('2/2/2020', periods=len(car_sales))
car_sales
```

```python
car_sales.plot('Sale Date', 'Total sales ($)')
plt.show()
```

```python
plt.scatter('Odometer (KM)', 'Price', data=car_sales)
plt.show()
```

###    New Data Set 

```python
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease.head(10)
```

```python
# histogram of the age 
heart_disease.hist(column='age', bins=50)
plt.show()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10, 8))
plt.show()
```

```python
over_age_50 = heart_disease[heart_disease['age'] > 50]
over_age_50.head(10)
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim([33, 80])
scatter = ax.scatter(x='age', y='chol',
                     data=heart_disease, 
                    c='target', cmap='jet')
ax.set(title='Heart Disease and Cholesterol levels', 
      xlabel='Age', 
      ylabel='Cholesterol')
ax.axhline(heart_disease['chol'].mean(), linestyle='--', color='green')
ax.legend(*scatter.legend_elements(), title='target')
plt.show()
```
